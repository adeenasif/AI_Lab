```python
# Task a
import math

# -------- Environment --------
class Environment:
    def __init__(self):
        self.game_tree = {
            'Start': ['ZoneA', 'ZoneB'],
            'ZoneA': ['A1', 'A2'],
            'ZoneB': ['B1', 'B2'],
            'A1': [], 'A2': [], 'B1': [], 'B2': []
        }

        self.leaf_scores = {
            'A1': 3, 'A2': 5,
            'B1': 2, 'B2': 9
        }

    def get_actions(self, state):
        return self.game_tree.get(state, [])

    def is_terminal(self, state):
        return len(self.get_actions(state)) == 0

    def evaluate(self, state):
        return self.leaf_scores.get(state, 0)


# -------- Agent --------
class RobotAgent:
    def __init__(self, env):
        self.env = env

    def minimax(self, state, is_max):
        print(f"Visiting State: {state} | Player: {'MAX (Robot)' if is_max else 'MIN (Thief)'}")

        # Terminal state
        if self.env.is_terminal(state):
            value = self.env.evaluate(state)
            print(f"Reached leaf {state} with value {value}\n")
            return value

        actions = self.env.get_actions(state)

        if is_max:
            best = -math.inf
            for action in actions:
                print(f"Robot considering move → {action}")
                val = self.minimax(action, False)
                best = max(best, val)
            print(f"Best value at {state} (MAX) = {best}\n")
            return best
        else:
            best = math.inf
            for action in actions:
                print(f"Thief considering move → {action}")
                val = self.minimax(action, True)
                best = min(best, val)
            print(f"Best value at {state} (MIN) = {best}\n")
            return best


# -------- Run --------
env = Environment()
agent = RobotAgent(env)

result = agent.minimax('Start', True)
print("Final Decision Value:", result)
```

```python
#Task 2
# part a

import math

game_tree = {
    'Start': ['Stop', 'Go', 'Turn'],   # root node 
    'Stop': ['E', 'F'],
    'Go': ['G', 'H'],
    'Turn': ['I', 'J'],
    'E': [], 'F': [], 'G': [], 'H': [], 'I': [], 'J': []  # leaves
}

# Scores for leaf nodes 
leaf_scores = { 'E': 6, 'F': 7, 'G': 2, 'H': 4, 'I': 8, 'J': 3 }

def get_nodes(node):
    return (game_tree.get(node, []))

def evaluate(node):
    return (leaf_scores.get(node, 0))

def minimax(node, depth, is_maximizing):
    children = get_nodes(node)

    if depth == 0 or len(children) == 0:
        return evaluate(node)
    
    if is_maximizing:
        best = -math.inf
        for c in children:
            best = max(best, minimax(c, depth-1, False))
        return (best)
    else:
        best = math.inf
        for c in children:
            best = min(best, minimax(c, depth-1, True))
        return best
    
result = minimax('Start', 2, True)
print("Minimax Best Score: ", result)
```
