```python
# Task 1
# A company’s security system consists of nine critical components (A through I). Each component can be either safe or vulnerable due to security flaws. A security agent is
# responsible for scanning the entire system, identifying vulnerabilities, and patching them to prevent attacks.
# You are tasked with simulating a cybersecurity exercise using the given environment and agent. Complete the following steps:

# Initial System Check: The system environment is initialized with random vulnerabilities. Display the initial state of the system, showing which components are safe and which are vulnerable.
import random

class Environment:
    def __init__(self):   # initialize all components with random safe or vulnerable states
        self.components = {
            'A': random.choice(['Safe', 'Vulnerable']),
            'B': random.choice(['Safe', 'Vulnerable']),
            'C': random.choice(['Safe', 'Vulnerable']),
            'D': random.choice(['Safe', 'Vulnerable']),
            'E': random.choice(['Safe', 'Vulnerable']),
            'F': random.choice(['Safe', 'Vulnerable']),
            'G': random.choice(['Safe', 'Vulnerable']),
            'H': random.choice(['Safe', 'Vulnerable']),
            'I': random.choice(['Safe', 'Vulnerable'])
        }

    def displayInitalState(self):   # display the initial state of each component    
        for comp, status in self.components.items():
            print(f"Component: {comp} | Status: {status}")
    
    def get_percept(self, component):  # return the current state of a component as a percept
        return self.components[component]
            
    def patch(self, component):  # patch the vulnerable component and mark it as safe
        self.components[component] = 'Safe'
        
class SimpleReflexAgent:
    def __init__(self):
        self.to_patch = []
    
    # System Scan: The security agent scans each component. If a component is vulnerable, the agent logs a warning and adds it to a list for patching. If it is secure, a success message is logged.

    def act(self, percept, component, environment): # take action based on the percept of the component
        if percept == 'Vulnerable':
            environment.patch(component)
            return f"WARNING: Component {component} was vulnerable → patched."
        else:
            return f"SUCCESS: Component {component} is already safe."

    # Patching Vulnerabilities: After the scan, the agent patches all vulnerable components, marking them as safe. Display messages indicating which components have been patched.

    def patch_vulnerabilities(self, environment):  # patch all components stored in the vulnerable list
        print("\nPatching Vulnerabilities...")
        for comp in self.to_patch:
            environment.patch(comp)
            print(f"Component {comp} patched successfully")
    
# Final System Check: Display the system’s final state, confirming that all vulnerabilities have been patched.

def run_agent(agent, environment):
    print("Initial System State:")
    for comp, state in environment.components.items():
        print(f"Component {comp}: {state}")

    print("\nSystem Scan & Actions:")
    for component in environment.components:
        percept = environment.get_percept(component)
        action = agent.act(percept, component, environment)
        print(action)

    print("\nFinal System State:")
    for comp, state in environment.components.items():
        print(f"Component {comp}: {state}")


agent = SimpleReflexAgent()
environment = Environment()

run_agent(agent, environment)
```

```python
# Task 2
# A data center runs multiple servers, each hosting different services. These services are either underloaded, balanced, or overloaded based on the system's load. 
# The Load Balancer Agent is responsible for redistributing tasks across the servers to ensure optimal system performance.

import random

class Environment:
    def __init__(self):
        self.servers = {}
        for i in range(5):  #  Create a system with 5 servers.
            self.servers[i+1] = random.choice(['Underloaded', 'Balanced', 'Overloaded']) # Each server has a load state of "Underloaded", "Balanced", or "Overloaded".

    def display_initial_state(self):
        # display the current load state of each server
        for server, load in self.servers.items():
            print(f"{server}: {load}")

    def get_percept(self, server):
        # return the load state of the server as a percept
        return self.servers[server]

    def redistribute_task(self, from_server, to_server):
        # move a task from an overloaded server to an underloaded server
        self.servers[from_server] = 'balanced'
        self.servers[to_server] = 'balanced'

# The Load Balancer Agent must scan the system and move tasks from overloaded servers to underloaded ones to balance the load.

class LoadBalancerAgent: #  Model Based
    def __init__(self):
        # maintain a model of the current server loads
        self.model = {}

    def update_model(self, server, percept):
        # update the internal model with the latest percept
        self.model[server] = percept

    def act(self, server, percept, environment):
        # update the model first
        self.update_model(server, percept)

        # if server is overloaded, find an underloaded server in the model
        if percept == 'overloaded':
            for target_server, target_load in self.model.items():
                if target_load == 'underloaded':
                    # redistribute task in the environment
                    environment.redistribute_task(server, target_server)
                    # update the model after redistribution
                    self.model[server] = 'balanced'
                    self.model[target_server] = 'balanced'
                    return f"moved tasks from {server} to {target_server} to balance load"
            return f"{server} is overloaded but no underloaded server available"
        else:
            return f"{server} load is {percept}, no action needed"


def run_agent(agent, environment): 
    print("initial server loads:")      # display initial system state
    environment.display_initial_state()

    print("\nload balancer actions:")
    for server in environment.servers:
        percept = environment.get_percept(server)
        action = agent.act(server, percept, environment)
        print(action)

    # display final system state after load balancing
    print("\nfinal server loads:")
    environment.display_initial_state()


agent = LoadBalancerAgent()
environment = Environment()

run_agent(agent, environment)
```

```python
# Task 3
import random

class Environment:
    def __init__(self):
        self.Task = {}      # either 'Failed' or 'Completed'
        for i in range(4):  # 4 tasks assumed
            self.Task[i + 1] = random.choice(['Failed', 'Completed'])

    def get_percept(self, index):
        return self.Task[index]

    def updateTask(self, index, new_state):  # update task with new status
        self.Task[index] = new_state

    def display(self, title):
        print(f"\n{title}")
        for task, status in self.Task.items():
            print(f"Task {task}: {status}")
        print("\n")

class BackupManagementAgent:
    def __init__(self):
        pass 

    def act(self, percept):
        if percept == 'Failed':
            return 'Completed'      # if backup is successful on first try
        else:
            return 'Task Already Successfully Completed'

    # run_agent is now a method of the Agent
    def run_agent(self, environment):
        environment.display('Initial Task Status')

        for i in range(1, 5): 
            percept = environment.get_percept(i)
            action = self.act(percept)
            print(f"Step {i}: Percept: {percept}, Action: {action}")

            if action == 'Completed':
                environment.updateTask(i, action)

        environment.display('Final Task Status')


# -------- MAIN --------
environment = Environment()
agent = BackupManagementAgent()
agent.run_agent(environment)
```

```python
# Task 4

import random

class Environment:
    def __init__(self):
        self.components = {}
        for i in range(9):
            self.components[chr(ord('A') + i)] = random.choice(['Safe', 'Low Risk Vulnerable', 'High Risk Vulnerable'])

    def get_percept(self):
        return self.components.copy()

    def display(self, title):
        print(f"\n{title}")
        for comp, status in self.components.items():
            print(f"Component {comp}: {status}")

    def update_component(self, component, new_state):
        self.components[component] = new_state


class CyberSecurityAgent:
    def __init__(self):
        self.utility_scores = {'Safe': 0, 'Low Risk Vulnerable': 10, 'High Risk Vulnerable': -5}

    def utility(self, state):
        return self.utility_scores[state]

    def act(self, percept):
        best_component = None
        best_utility = -float('inf')

        for comp, state in percept.items():
            score = self.utility(state)
            if score > best_utility:
                best_utility = score
                best_component = comp

        state = percept[best_component]

        if state == 'Low Risk Vulnerable':
            action = 'Patch'
        elif state == 'High Risk Vulnerable':
            action = 'Requires Premium Service'
        else:
            action = 'No Action Needed'

        return best_component, action

    def run_agent(self, environment):
        environment.display("Initial System Status")
        print("")

        # First pass: Patch all Low Risk Vulnerable components
        while True:
            percept = environment.get_percept()
            if all(state != 'Low Risk Vulnerable' for state in percept.values()):
                break

            component, action = self.act(percept)
            state = percept[component]

            if state == 'Low Risk Vulnerable':
                print(f"Warning: Component {component} is LOW RISK VULNERABLE. Action: {action}.")
                environment.update_component(component, 'Safe')
                print(f"Component {component} patched successfully.")
            elif state == 'Safe':
                print(f"Component {component} is SAFE. No action needed.")
            elif state == 'High Risk Vulnerable':
                print(f"Warning: Component {component} is HIGH RISK VULNERABLE. Action: {action}")

        print("\nLow Risk Vulnerability HANDLED...now High Risk Vulnerability being checked\n")
        # Second pass: Log all High Risk components without patching
        percept = environment.get_percept()
        for component, state in percept.items():
            if state == 'High Risk Vulnerable':
                print(f"Warning: Component {component} is HIGH RISK VULNERABLE. Action: {action}")
            elif state == 'Safe':
                print(f"Component {component} is SAFE. No action needed.")

        environment.display("Final System Status")

env = Environment()
agent = CyberSecurityAgent()
agent.run_agent(env)
```

```python
# Task 5

import random

class Environment:
    def __init__(self):
        # define locations in the hospital
        self.rooms = [f'room{i+1}' for i in range(5)]
        self.nurse_stations = ['nurse_station1', 'nurse_station2']
        self.medicine_storage = ['storage1', 'storage2']

        # generate a schedule: which patient in which room needs which medicine
        self.schedule = {
            room: random.choice(['medA', 'medB', 'medC'])
            for room in self.rooms
        }

        self.deliveries_done = {room: False for room in self.rooms}   # keep track of deliveries

    def display_schedule(self):         # display scheduled medicines for each room
        print("Medicine delivery schedule:")
        for room, med in self.schedule.items():
            print(f"{room}: {med}")

    def check_delivery(self, room):     # return whether the delivery for a room is done
        return self.deliveries_done[room]

    def mark_delivered(self, room):     # mark a delivery as completed
        self.deliveries_done[room] = True

    def staff_available(self):          # randomly decide if staff is available to assist
        return random.choice([True, False])


class GoalBasedDeliveryRobot:
    def __init__(self, environment):
        self.env = environment
        self.location = 'storage1'
        self.inventory = None

    def move_to(self, location):           # move robot to the specified location
        self.location = location
        print(f"Moving to {location}")

    def pick_up_medicine(self, medicine):  # pick up medicine from storage
        self.inventory = medicine
        print(f"Picked up {medicine}")

    def scan_patient_id(self, room):       # simulate scanning patient id
        print(f"Scanned patient id in {room}")

    def alert_staff(self, room):           # alert nurses or doctors
        print(f"Alerting staff for assistance in {room}")

    def deliver_medicine(self, room):      # goal-based decision: deliver medicine to the room
        if self.inventory is None:
            print(f"No medicine to deliver to {room}")
            return

        self.move_to(room)
        self.scan_patient_id(room)

        # randomly simulate a critical situation
        if not self.env.staff_available():
            self.alert_staff(room)
            print(f"Cannot deliver medicine to {room} without staff")
            return

        print(f"Delivered {self.inventory} to {room}")
        self.env.mark_delivered(room)
        self.inventory = None

    def execute_delivery_plan(self):    # go through all scheduled deliveries and achieve the goal
        for room, medicine in self.env.schedule.items():
            if not self.env.check_delivery(room):
                self.move_to(random.choice(self.env.medicine_storage))
                self.pick_up_medicine(medicine)
                self.deliver_medicine(room)

env = Environment()
robot = GoalBasedDeliveryRobot(env)

# display initial schedule
env.display_schedule()

# execute deliveries (goal-based)
print("\nRobot executing delivery plan:\n")
robot.execute_delivery_plan()

# final check
print("\nFinal delivery status:")
for room, done in env.deliveries_done.items():
    status = 'completed' if done else 'pending'
    print(f"{room}: {status}")
```

```python
# Task 6

class Environment:
    def __init__(self):
        # initialize the rooms with their status
        # rooms 'c', 'e', 'j' have fires, others are safe
        self.rooms = {
            'a': 'safe',
            'b': 'safe',
            'c': 'fire',
            'd': 'safe',
            'e': 'fire',
            'f': 'safe',
            'g': 'safe',
            'h': 'safe',
            'j': 'fire'
        }

    def display_environment(self): # display the environment in a 3x3 grid
        # using 'F' for fire, 'S' for safe
        grid_order = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'j']
        print("\nEnvironment status:")
        for i, room in enumerate(grid_order):
            status_symbol = 'F' if self.rooms[room] == 'fire' else 'S'
            print(status_symbol, end=' ')
            if (i + 1) % 3 == 0:
                print()  # new line for grid row

    def extinguish_fire(self, room):
        # change room status from fire to safe
        if self.rooms[room] == 'fire':
            self.rooms[room] = 'safe'
            print(f"Fire detected in {room}! extinguishing...")
        else:
            print(f"{room} is safe, moving on.")


class FirefightingRobot:
    def __init__(self, environment):
        # robot knows the environment and its predefined path
        self.env = environment
        self.path = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'j']
        self.current_room = 'a'

    def move_and_extinguish(self):
        # move through the path and handle fires
        for room in self.path:
            self.current_room = room
            print(f"\nRobot moving to {room}")
            self.env.extinguish_fire(room)
            self.env.display_environment()



env = Environment()
robot = FirefightingRobot(env)

print("initial environment:")
env.display_environment()

robot.move_and_extinguish()

print("\nfinal environment after robot completed movement:")
env.display_environment()
```
